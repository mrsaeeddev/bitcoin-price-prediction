```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as dateconv
import datetime as dt

df = pd.read_csv('bitcoin.csv')

df.head()
```

```python
df.describe()
```

```python
print(df.count())
```

```python
df['Months'] = pd.to_datetime(df['Date']).dt.month.astype(np.uint8)

features = ['Open','High','Low','Close']

target = 'MarketCap'

X = df[features]

Y = df[target]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.50, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)

predictions = lr.predict(X_test)
```

```python
# plt.scatter(X_train, y_train, color="red")
plt.plot(X_train, y_train, 'o--', color="blue")

plt.plot(X_test, predictions, 'o--', color="green")
plt.title('Features vs Prediction(Testing  and Training)')
plt.xlabel('Features')
plt.ylabel('Target')
```
